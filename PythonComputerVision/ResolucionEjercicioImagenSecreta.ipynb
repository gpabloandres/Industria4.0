Explorar los canales de color de la imagen BGR

```python
img_bgr = cv2.imread("/imagenSecreta.jpeg", cv2.IMREAD_COLOR)
b, g, r = cv2.split(img_bgr)
plt.figure(figsize=(15, 5))

plt.subplot(142)
plt.imshow(r, cmap="OrRd");plt.title("Rojos")
plt.subplot(143)
plt.imshow(g, cmap="YlGn");plt.title("Verdes")
plt.subplot(144)
plt.imshow(b, cmap="Blues");plt.title("Azules")
```

Combinar los tres canales nuevamente

```python
imgCombinada = cv2.merge((b, g, r))
img_rgb = cv2.cvtColor(imgCombinada, cv2.COLOR_BGR2RGB)

plt.subplot(141)
plt.imshow(img_rgb)
plt.tight_layout()
plt.title("Canales combinados")
plt.show()
```

Combinar el canal BLUE + GREEN

```python
img_BGCombinada = cv2.merge((b, g, np.zeros_like(r)))
img_gb = cv2.cvtColor(img_BGCombinada, cv2.COLOR_BGR2RGB)

plt.subplot(141)
plt.imshow(img_gb)
plt.tight_layout()
plt.title("Canales BG combinados")
plt.show()
```

Combinar el canal GREEN + RED

```python
img_GRCombinada = cv2.merge((g, r, np.zeros_like(b)))
img_gr = cv2.cvtColor(img_GRCombinada, cv2.COLOR_BGR2RGB)

plt.subplot(141)
plt.imshow(img_gr)
plt.tight_layout()
plt.title("Canales GR combinados")
plt.show()
```
